```python
def trans_df(df):
    df = df[['neighbor', '1', '2', '3']].dropna()
    df['1'] = df['1'].tolist()
    df['2'] = df['2'].tolist()
    df['3'] = df['3'].tolist()
    df['neighbor'] = df['neighbor'].tolist()
    return df
data_2004 = pd.read_csv('2004_zip_result.csv')
data_2004['neighbor']
#data_2004['type'] = data_2010['1']+data_2010['2']+data_2010['3']+data_2010['4']+data_2010['5']
data_2004 = trans_df(data_2004)
data_2004
```

```python
def get_type_list(dataframe):
    type_list = []
    type_column_1 = dataframe['1']
    type_column_2 = dataframe['2']
    type_column_3 = dataframe['3']
    #print(type_column)
    columns = [type_column_1,type_column_2,type_column_3]
    for column in columns:
        for row in column:
            row = row.replace('[','')
            row = row.replace(']', '')
            row = row.strip().split(',')
            #print(row)
            if row[0] not in type_list:
                type_list.append(row[0])

    #print(type_list)
    new_df = pd.DataFrame(columns = type_list)
    new_df['neighbor'] = dataframe['neighbor']
    #print(new_df)
    return new_df, type_list

new_df, type_list = get_type_list(data_2004)
#print(new_df.iat[1,2])
```

```python
def final_df(dataframe):
    new_df,type_list = get_type_list(dataframe)
    for i in range(30):
        for data in dataframe.iloc[i][1:]:
            #print(data)
            data = data.replace('[','')
            data = data.replace(']', '')
            data = data.strip().split(',')
            index = type_list.index(data[0])
            #print(data[1])
            #print(new_df.iat[i,index])
            new_df.iat[i,index] = data[1]
    return new_df
```

```python
print("2004 summary")
new_df = final_df(data_2004)
new_df
```

```python
new_df.iplot(kind='bar', xTitle='neighbor', yTitle='number of complaint type',    title='complaint type with borough', colorscale='set2')
```
